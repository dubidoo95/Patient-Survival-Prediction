```python
# age, bmi, weight, height에 결측치 있는 row 삭제
df = df[(df.age >= 1) & (df.bmi >= 1) & (df.weight >= 1) & (df.height >= 1)]
df.reset_index(drop=True, inplace=True)

# age, height, weight column의 dtype을 int로 변환
df = df.astype({"age": "int64","height": "int64", "weight": "int64"})
```

```python
# 숫자형 data들을 모두 int형으로 전환
def to_int(df):
    for column in df.columns:
        if df[column].dtype=='float64':
            df[column] = df[column].fillna(-1000)
            df[column] = df[column].astype('int64')
            df[column] = df[column].replace(-1000, np.nan)
    return df

df = to_int(df)
df.info()
```

```python
df.nunique()
```

```python
# nunique가 100 넘는(cardinality가 큰) column 삭제
for column in df.columns:
    if df[column].nunique() > 100:
        df.drop(column, axis=1, inplace=True)
df.reset_index(drop=True, inplace=True)
df.shape
```

```python
# target, features 지정
target = 'hospital_death'
features = df.drop(target, axis=1).columns

# train, val, test set 분리
train, test = train_test_split(df, test_size=0.2, stratify=df[target])
train, val = train_test_split(train, test_size=0.2, stratify=train[target])

# features, target 분리
X_train, y_train = train[features], train[target]
X_val, y_val = val[features], val[target]
X_test, y_test = test[features], test[target]

# 제대로 분리되었는지 확인
print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_val: {X_val.shape}, y_val: {y_val.shape}")
print(f"X_test: {X_test.shape}, y_test: {y_test.shape}")
```

```python
# 기준 모델
base = y_train.mode()[0]
baseline = [base]*len(y_train)
print(f"기준 모델의 정확도: \n{classification_report(y_train, baseline)}")
```

```python
# decision tree
dt = make_pipeline(
    OrdinalEncoder(),
    SimpleImputer(),
    DecisionTreeClassifier(
        criterion="entropy",
        class_weight="balanced",
        random_state=42))

# 모델 학습
dt.fit(X_train, y_train)
y_val_pred = dt.predict(X_val)
print(f"desicion tree 모델 결과: \n{classification_report(y_val, y_val_pred)}")
```

```python
# randomforest model
rf = make_pipeline(
    OrdinalEncoder(),
    SimpleImputer(),
    RandomForestClassifier(
        criterion="entropy",
        oob_score=True,
        n_jobs=-1,
        random_state=42,
        class_weight='balanced'))

# 모델 학습
rf.fit(X_train, y_train)
y_val_pred = rf.predict(X_val)
print(f"randomforest 모델 결과: \n{classification_report(y_val, y_val_pred)}")
```

```python
# xgbclassifier 모델 작성
xgb = make_pipeline(
    OrdinalEncoder(),
    SimpleImputer(strategy='mean'),
    XGBClassifier(
        objective="binary:logistic",
        n_jobs=-1,
        eval_metric='auc',
        learning_rate=0.05,
        scale_pos_weight=(y_train==0).sum()/(y_train==1).sum(),
        reg_lambda=1))

# 모델 학습
xgb.fit(X_train, y_train)
y_val_pred = xgb.predict(X_val)
print(f"xgb 모델 결과: \n{classification_report(y_val, y_val_pred)}")
```

decision tree, randomforestclassifier, xgbclassifier를 사용하여 각각 모델을 만들어 본 결과 xgbclassifier의 성능이 가장 좋았기 때문에 xgbclassifier를 사용하여 이후 학습 진행

```python
# 검증할 파라미터들 작성
params = {
    "xgbclassifier__max_depth": [4, 6, 8, 10],
    "xgbclassifier__n_estimators": [400, 500, 600],
    "xgbclassifier__min_child_weight": [1, 2, 4, 8],
    "xgbclassifier__colsample_bytree": [0.6, 0.7, 0.8, 0.9],
    "xgbclassifier__subsample": [0.6, 0.7, 0.8, 0.9]
}

# randomized search로 hyperparameters 탐색
randomized_search = RandomizedSearchCV(
    xgb,
    param_distributions=params,
    n_iter=50,
    scoring="f1",
    cv=3,
    verbose=3,
    random_state=42)

randomized_search.fit(X_train, y_train)
```

```python
# randomized search로 구한 hyperparameter로 model 작성
pipeline = randomized_search.best_estimator_
pipeline.fit(X_train, y_train)
y_val_pred = pipeline.predict(X_val)
print(f"hyperparametes 조정 후 결과: \n{classification_report(y_val, y_val_pred)}")
```

```python
# permuter 정의
permuter = PermutationImportance(
    pipeline.named_steps["xgbclassifier"],
    scoring="f1",
    n_iter=20,
    random_state=2)

# 스코어 다시 계산
X_test_preprocessed = pipeline[0].transform(X_test)
permuter.fit(X_test_preprocessed, y_test)
```

```python
# importance가 높은 순서대로 feature 정렬 및 시각화
feature_names = X_test.columns.tolist()
pi = pd.Series(permuter.feature_importances_, feature_names).sort_values()
plt.figure(figsize=(10,20))
pi.plot.barh()
plt.title("Permutation Importance")
```

```python
# 특성별 score 확인
eli5.show_weights(
    permuter,
    top=None,
    feature_names=feature_names)
```

```python
def get_roc_auc(model, X_val_encoded, y_val):
    if isinstance(model, XGBClassifier):
        y_pred_proba = model.predict_proba(
            X_val_encoded, iteration_range=(0, model.best_iteration)
        )[:, 1]
    else:
        y_pred_proba = model.predict_proba(X_val_encoded)[:, 1]

    print("ROC-AUC:", roc_auc_score(y_val, y_pred_proba))

    baseline_fpr, baseline_tpr, _ = roc_curve(y_val, [0] * len(y_val))
    fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
    tmp = pd.DataFrame({'fpr':fpr, 'tpr':tpr, 'thresholds': thresholds})
    plt.plot(
        baseline_fpr, baseline_tpr, linestyle="--", color="green", label="Baseline"
    )
    plt.plot(fpr, tpr, color="blue", label="Model")
    plt.scatter(fpr, tpr, color="red")
    plt.legend()
    plt.title("ROC curve")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.show()
    return tmp
def summarize_results(model, X_val_encoded, y_val):
    # 1. Train / Val Loss Plot
    if isinstance(model, XGBClassifier):
        results = model.evals_result()
        tgts = results["validation_0"].keys()

        for tgt in tgts:
            train_error = results["validation_0"][tgt]
            val_error = results["validation_1"][tgt]

            plt.plot(train_error, label="Train")
            plt.plot(val_error, label="Validation")
            plt.ylabel(tgt)
            plt.xlabel("Model Complexity (n_estimators)")
            plt.legend()
            plt.show()

    # 2. Accuracy / Classification Report
    print("검증 정확도", model.score(X_val_encoded, y_val))

    if isinstance(model, XGBClassifier):
        y_pred = model.predict(X_val_encoded, iteration_range=(0, model.best_iteration))
    else:
        y_pred = model.predict(X_val_encoded)

    print(classification_report(y_val, y_pred))

    # 3. ROC-AUC
    get_roc_auc(model, X_val_encoded, y_val)

    # 4. Confusion Matrix
    ConfusionMatrixDisplay.from_estimator(model, X_val_encoded, y_val,cmap='Blues')
    plt.show()
    
summarize_results(model, X_val_imputed, y_val)
```

```python
def get_ice_plot_data(data, data_index, target_feature, target_feature_range):
    one_datapoint = data.loc[[data_index]].copy()
    results = []
    
    for target_value in target_feature_range:
        one_datapoint[target_feature] = target_value
        one_datapoint_pred_proba = model.predict_proba(one_datapoint)[:,1]
        results.append(one_datapoint_pred_proba.item())
    results = np.array(results)
    return target_feature_range, results-results[0]

# target_feature에 따른 ICE plot
base_index = X_test[(X_test.age==X_test.age.min()) & (X_test.bmi>=18) & (X_test.bmi<=25) & (X_test.ventilated_apache==0)].index
target_features = ["age", "bmi", "ventilated_apache"]
for target_feature in target_features:   
    target_feature_range = range(X_test_encoded[target_feature].min(), X_test_encoded[target_feature].max() + 1)
    plt.plot(*get_ice_plot_data(
                X_test_encoded, base_index[0], target_feature, target_feature_range))
    plt.title(f'pred_proba change by {target_feature}')
    plt.xlabel(target_feature)
    plt.ylabel("$\Delta$ pred_proba")
    plt.show()
```
